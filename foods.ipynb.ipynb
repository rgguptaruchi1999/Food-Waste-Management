```python
import streamlit as st
import pandas as pd
import sqlite3

# Assuming the run_food_wastage_queries function is defined here or imported
def run_food_wastage_queries(db_path="food_wastage.db", city_name='New Jessica'):
    """
    Connects to the food wastage database, runs a series of analytical queries,
    and returns the results as a dictionary of dataframes.

    Args:
        db_path (str): The path to the SQLite database file.
        city_name (str): The city name to use for the parameterized query.

    Returns:
        dict: A dictionary where keys are query names (e.g., 'q1', 'q2') and
              values are the resulting pandas DataFrames, or None if an error occurs.
    """
    conn = None  # Initialize connection to None
    query_results = {}

    try:
        # Connect to DB
        conn = sqlite3.connect(db_path)
        cursor = conn.cursor()

        # Insert data from DataFrames into SQLite tables in "food_wastage.db"
        # Use if_exists='replace' to handle table creation and data insertion
        # Assuming the dataframes (providers, receivers, food_listings, claims) are available in the environment
        # If not, you would need to load them from CSV files here.
        try:
            # Example of loading dataframes if they are not already in memory
            providers = pd.read_csv('providers_data_clean.csv')
            receivers = pd.read_csv('receivers_data_clean.csv')
            food_listings = pd.read_csv('food_listings_data_clean.csv')
            claims = pd.read_csv('claims_data_clean.csv')

            providers.to_sql('providers', conn, if_exists='replace', index=False)
            receivers.to_sql('receivers', conn, if_exists='replace', index=False)
            food_listings.to_sql('food_listings', conn, if_exists='replace', index=False)
            claims.to_sql('claims', conn, if_exists='replace', index=False)

            print("Data inserted into tables successfully in food_wastage.db")
        except Exception as e:
            print(f"Error inserting data into tables: {e}")


        # 1. Number of food providers per city
        q1 = """
        SELECT
            City,
            COUNT(Provider_ID) AS no_of_food_providers
        FROM providers
        GROUP BY City
        ORDER BY no_of_food_providers DESC;
        """

        # 2. Food provider type contributing the most food
        q2 = """
        SELECT
            Type AS Provider_Type,
            SUM(f.Quantity) AS total_quantity
        FROM food_listings f
        JOIN providers p ON f.Provider_ID = p.Provider_ID
        GROUP BY Type
        ORDER BY total_quantity DESC;
        """

        # 3. Contact info of food providers in a specific city (parameterized)
        q3 = """
        SELECT DISTINCT
            p.Provider_ID,
            p.Name,
            p.Contact,
            p.Address,
            p.City,
            f.Food_Name,
            f.Quantity
        FROM providers p
        JOIN food_listings f ON p.Provider_ID = f.Provider_ID
        WHERE p.City = ?;
        """

        # 4. Total quantity claimed per provider
        q4 = """
        SELECT
            p.Provider_ID,
            p.Name AS Provider_Name,
            SUM(f.Quantity) AS total_quantity_claimed
        FROM claims c
        JOIN food_listings f ON c.Food_ID = f.Food_ID
        JOIN providers p ON f.Provider_ID = p.Provider_ID
        WHERE c.Status = 'Completed'
        GROUP BY p.Provider_ID, p.Name
        ORDER BY total_quantity_claimed DESC;
        """

        # 5. Total food available per provider
        q5 = """
        SELECT
            p.Provider_ID,
            p.Name AS Provider_Name,
            SUM(f.Quantity) AS Total_Food_Quantity
        FROM providers p
        JOIN food_listings f ON p.Provider_ID = p.Provider_ID
        GROUP BY p.Provider_ID, p.Name
        ORDER BY Total_Food_Quantity DESC;
        """

        # 6. City with highest number of food listings
        q6 = """
        SELECT Location AS City, COUNT(*) AS total_food_listings
        FROM food_listings
        GROUP BY Location
        ORDER BY total_food_listings DESC
        LIMIT 1;
        """

        # 7. Most commonly available food types
        q7 = """
        SELECT Food_Type, COUNT(*) AS total_count
        FROM food_listings
        GROUP BY Food_Type
        ORDER BY total_count DESC
        LIMIT 1;
        """

        # 8. Number of claims per food item
        q8 = """
        SELECT f.Food_Name, COUNT(*) AS total_claims
        FROM food_listings f
        JOIN claims c ON f.Food_ID = c.Food_ID
        GROUP BY f.Food_Name
        ORDER BY total_claims DESC;
        """

        # 9. Provider with highest number of successful claims
        q9 = """
        SELECT p.Provider_ID, p.Name, COUNT(*) AS successful_claims
        FROM claims c
        JOIN food_listings f ON f.Food_ID = c.Food_ID
        JOIN providers p ON f.Provider_ID = p.Provider_ID
        WHERE c.Status = 'Completed'
        GROUP BY p.Provider_ID, p.Name
        ORDER BY successful_claims DESC;
        """

        # 10. Percentage of claims by status
        q10 = """
        SELECT
            Status,
            COUNT(*) AS total_claims,
            ROUND((COUNT(*) * 100.0 / (SELECT COUNT(*) FROM claims)), 2) AS percentage
        FROM claims
        GROUP BY Status;
        """

        # 11. Average quantity claimed per food item
        q11 = """
        SELECT ROUND(AVG(f.Quantity), 2) AS avg_quantity_claimed
        FROM claims c
        JOIN food_listings f ON c.Food_ID = c.Food_ID;
        """

        # 12. Most claimed meal type
        q12 = """
        SELECT f.Meal_Type, COUNT(*) AS total_claims
        FROM claims c
        JOIN food_listings f ON c.Food_ID = f.Food_ID
        GROUP BY f.Meal_Type
        ORDER BY total_claims DESC;
        """

        # 13. Total food donated per provider
        q13 = """
        SELECT p.Name, SUM(f.Quantity) AS total_quantity_donated
        FROM food_listings f
        JOIN providers p ON f.Provider_ID = p.Provider_ID
        GROUP BY p.Name
        ORDER BY total_quantity_donated DESC;
        """

        # 14. Meal type donated vs claimed
        q14 = """
        SELECT
            f.Meal_Type,
            COUNT(DISTINCT f.Food_ID) AS total_donated,
            SUM(CASE WHEN c.Status = 'Completed' THEN 1 ELSE 0 END) AS total_claimed
        FROM food_listings f
        LEFT JOIN claims c ON f.Food_ID = c.Food_ID
        GROUP BY f.Meal_Type
        ORDER BY total_donated DESC;
        """

        # 15. Day of week with most donations
        q15 = """
        SELECT
            CASE strftime('%w', Expiry_Date)
                WHEN '0' THEN 'Sunday'
                WHEN '1' THEN 'Monday'
                WHEN '2' THEN 'Tuesday'
                WHEN '3' THEN 'Wednesday'
                WHEN '4' THEN 'Thursday'
                WHEN '5' THEN 'Friday'
                WHEN '6' THEN 'Saturday'
            END AS donation_day_of_week,
            COUNT(*) AS total_donations
        FROM food_listings
        GROUP BY donation_day_of_week
        ORDER BY total_donations DESC;
        """

        # Execute all queries and store results
        queries = {
            'Number of food providers per city': (q1, None),
            'Food provider type contributing the most food': (q2, None),
            f'Contact info of food providers in {city_name}': (q3, (city_name,)),
            'Total quantity claimed per provider': (q4, None),
            'Total food available per provider': (q5, None),
            'City with highest number of food listings': (q6, None),
            'Most commonly available food types': (q7, None),
            'Number of claims per food item': (q8, None),
            'Provider with highest number of successful claims': (q9, None),
            'Percentage of claims by status': (q10, None),
            'Average quantity claimed per food item': (q11, None),
            'Most claimed meal type': (q12, None),
            'Total food donated per provider': (q13, None),
            'Meal type donated vs claimed': (q14, None),
            'Day of week with most donations': (q15, None)
        }

        for query_name, (q, params) in queries.items():
            try:
                if params:
                    df = pd.read_sql_query(q, conn, params=params)
                else:
                    df = pd.read_sql_query(q, conn)
                query_results[query_name] = df
            except Exception as e:
                print(f"Error executing query '{query_name}': {e}")
                query_results[query_name] = pd.DataFrame() # Store empty dataframe on error

    except Exception as e:
        st.error(f"An error occurred during database operation: {e}")
        query_results = None # Indicate overall failure
    finally:
        if conn:
            conn.close()

    return query_results


st.title("Food Wastage Analysis")

# Call the function to get query results
query_results_dict = run_food_wastage_queries()

# Check if the function call was successful and results are available
if query_results_dict is not None and 'query_results' in st.session_state:
    st.success("Data loaded and queries executed successfully.")

    # Display each query result
    for query_name, result_df in st.session_state['query_results'].items():
        st.subheader(query_name)
        if not result_df.empty:
            st.write(f"Displaying results for '{query_name}':")
            st.dataframe(result_df)
        else:
            st.info(f"No results found for '{query_name}' or an error occurred.")

elif query_results_dict is None:
    st.error("Failed to load data or execute queries. Please check the console for details.")
else:
     st.info("Running queries...") # Initial state before results are available
```
